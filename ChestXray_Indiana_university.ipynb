```python
# Colab Notebook for Exploring Indiana Chest X-ray Reports CSV

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Upload indiana_reports.csv from your machine
from google.colab import files
uploaded = files.upload()

# Load the dataset
csv_filename = list(uploaded.keys())[0]
df = pd.read_csv(csv_filename)

print("✅ Loaded Dataset:")
print(f"Shape (rows, columns): {df.shape}\n")

# Show general info
print("📌 DataFrame Info:")
df.info()

print("\n📌 Statistical Summary (if numeric columns present):")
print(df.describe(include='all'))

print("\n📌 Columns:")
print(df.columns.tolist())

print("\n📌 First Few Rows:")
display(df.head(5))

# Display sample values for key columns
key_cols = ['uid', 'MeSH', 'Problems', 'indication', 'comparison', 'findings', 'impression']
for col in key_cols:
    if col in df.columns:
        print(f"\n🔹 Unique Value Samples for '{col}':")
        display(df[col].dropna().unique()[:5])

# Optional: visualize missing values
plt.figure(figsize=(10,6))
sns.heatmap(df.isna(), cbar=False)
plt.title("Missing Values in Dataset Columns")
plt.show()
```
